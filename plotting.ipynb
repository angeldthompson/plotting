```python
import pandas as pd
import matplotlib.pyplot as plt

heart = pd.read_csv('Medicaldataset.csv')

heart
```

```python
# Count the occurrences of each gender
gender_counts = heart['Gender'].value_counts()

# Plot the counts where 1 is for male and 0 is for female
gender_counts.plot(kind='bar')
```

```python
# Count the occurrences of each Result
result_counts = heart['Result'].value_counts()

# Plot the counts of whether the person had a heart attack or not
result_counts.plot(kind='bar')
```

```python
crime = pd.read_csv('Hate_Crimes_2017-2025.csv')
crime.head()
```

```python
crime.value_counts('Offense(s)')
```

```python
#Plot number of victims under 18
sns.histplot(crime['Number of Victims under 18'])
```

```python
import plotly.express as px

#Get counts for the different Offenses
offense_counts = crime['Offense(s)'].value_counts().reset_index()

#Rename columns for offense counts
offense_counts.columns = ['Offense(s)', 'Count']

#Create tree map
fig = px.treemap(offense_counts, path=['Offense(s)'], values='Count', title='Counts of Offense(s)')

#Plot treemap
fig.show()
```

```python
#Calculate value counts for day of the week
day_counts = crime['Day of Week'].value_counts()

#Plot pie chart
day_counts.plot(kind='pie')

#Add title and remove y-axis label
plt.title('Distribution of days per week offenses occurred')
plt.ylabel('')

#Display chart
plt.show()
```

```python
#Ensure column is in datetime format
crime['Date of Incident'] = pd.to_datetime(crime['Date of Incident'])

#Extract month and year
crime['Month'] = crime['Date of Incident'].dt.month
crime['Year'] = crime['Date of Incident'].dt.year

#Look at crime df
crime
```

```python
crime = crime.drop(['month', 'year'], axis=1)
```

```python
#Group by Year and Offenses
offense_over_time = crime.groupby(['Year', 'Offense(s)']).size().unstack(fill_value=0)

#Set size of plot
plt.figure(figsize=(24,16))


for offense in offense_over_time.columns:
    plt.plot(offense_over_time.index, offense_over_time[offense], label= offense)

#Set labels and title
plt.xlabel('Year')
plt.ylabel('Offense Count')
plt.title('Offense Counts Over Time')
plt.legend(title='Offense(s) Type')
plt.grid(True)
plt.tight_layout()

#Plot graph
plt.show()
```

```python
#Set size of plot
plt.figure(figsize=(8, 6))

#Plot Bias count and add labels
sns.countplot(data=crime, y='Bias')
plt.title('Count of Items in Each Category (Horizontal)')
plt.xlabel('Count')
plt.ylabel('Bias')

#Show plot of bias counts 
plt.show()
```
